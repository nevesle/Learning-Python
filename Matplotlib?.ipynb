```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fig, ax = plt.subplots() # fig is a container that holds  everything on the page , axes is the part of the page that holds the data i.e the canvas that is been drawn on
```

```python
train = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
train.head()
```

```python
train.plot(x = "total_rooms",y = "median_house_value", kind= "line")
```

```python
fig, ax = plt.subplots()
ax.plot(train["total_rooms"], train["median_house_value"], marker = "o", linestyle = "None", color = "y")
ax.set_xlabel("Total Rooms")
ax.set_ylabel("Median House Value")
ax.set_title("Total rooms  vs   Median House Value")
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(train["total_rooms"], train["median_house_value"], marker = "o", linestyle = "None", color = "y")
ax[1].plot(train["total_rooms"], train["median_house_value"], marker = "o", linestyle = "None", color = "y")
```
